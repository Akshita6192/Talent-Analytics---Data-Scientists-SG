## **Education Analysis**

```python
df["Education Degree 1"].isna().sum()
```

```python
df["Education Degree 1"].unique()
```

```python
def degree(deg):
    
    if 'bachelor' in deg.lower():
        return 'Bachelors'
    elif 'master' in deg.lower() or 'post graduation' in deg.lower() or "m.sc." in deg.lower() or 'ma' in deg.lower() or 'mba' in deg.lower() or 'mphil' in deg.lower() or 'm.s'in deg.lower() or 'msc' in deg.lower():
        return 'Masters'
    elif 'm soc sci' in deg.lower() or 'masters' in deg.lower() or 'ms' in deg.lower() or 'm.s.' in deg.lower():
        return 'Masters'
    elif 'doctor' in deg.lower():
        return 'Doctor/Ph.D'
    elif 'ph.d.' in deg.lower() or 'ph.d' in deg.lower() or 'phd' in deg.lower() or 'doctor' in deg.lower() :
        return 'Doctor/Ph.D'
    else:
        return 'Bachelors'
    


    
```

```python
df["degree1"] = df["Education Degree 1"].apply(degree)
df["degree1"].value_counts()
```

```python
plt.figure(figsize= (8,8))
plt.pie(df["degree1"].value_counts(), labels= ['Masters', 'Bachelor', 'Doctor or Ph.D'], 
        shadow= True, autopct= '%1.0f%%', explode = (0,0.1,0.1)
       )
plt.title("Highest Educational Qualification Split")
plt.show()
```

```python
def education(edu):
    if 'analytics' in edu.lower():
        return 'Analytics'
    if 'mathematics' in edu.lower():
        return 'Mathematics'
    if 'accountancy' in edu.lower() or 'accounting' in edu.lower():
        return 'Accountancy'
    if 'statistics' in edu.lower():
        return 'Statistics'
    if 'engineering' in edu.lower() or 'computer science' in edu.lower() or 'technology' in edu.lower() or 'information technology'in edu.lower():
        return 'Engineering/Technology'
    if 'biology' in edu.lower() or 'biomedical' in edu.lower() or 'biological' in edu.lower():
        return 'Biology/Biomedical'
    if 'administration' in edu.lower() or 'management' in edu.lower():
        return 'Management/administration'
    if 'data science' in edu.lower() or 'artificial' in edu.lower() or "machine learning" in edu.lower() or 'deep learning' in edu.lower():
        return "Data Science/AI/ML"
    if 'chemical' in edu.lower():
        return 'Chemical'
    if 'finance' in edu.lower():
        return 'Finance'
    if 'physics' in edu.lower():
        return 'Physics'
    else:
        return 'Others'
```

```python
df["Education Domain"] = df["Education FOS 1"].apply(education)
df["Education Domain"].value_counts()
```

```python
df["Education Domain"].value_counts().plot(kind ='bar', figsize = (10,8), title = "Educational Background Split", grid = True)
```

## Designation Split

### 1. Overall Designation split

```python
def title(tit):
    if 'chief' in tit.lower():
        return 'Chief Data Scientist'
    elif 'senior' in tit.lower():
        return 'Senior Data Scientist'
    elif 'lead' in tit.lower():
        return 'Lead Data Scientist'
    elif 'junior' in tit.lower() or 'intern' in tit.lower():
        return 'Data Scientist'
    elif 'avp' in tit.lower():
        return 'AVP'
    elif 'vp' in tit.lower() or 'vice president' in tit.lower():
        return 'VP'
    else:
        return 'Data Scientist'
    
df["Title_category"] = df["Title"].apply(title)
```

```python
df["Title_category"].value_counts()
```

```python
plt.figure(figsize= (8,8))
plt.pie(df["Title_category"].value_counts(), labels= ['Data Scientist', 'Senior Data Scientist', 'Lead Data Scientist','Chief Data Scientist', 'AVP','VP'], 
        shadow= True, autopct= '%1.0f%%', explode = (0,1,1,1,1,1)
       )
plt.show()
```

### Let's create a WordCloud for the Skills as listed by Data Scientists on LinkedIn

```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
```

```python
text = " ".join(review for review in df.Skills)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)


# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize = (12,12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
